```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#functional method
```

```python
plt.plot(x,y,'r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('visual-presentation')
plt.show()
```

```python
#subplots
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.show()

plt.subplot(1,2,2)
plt.plot(x,y,'r')
plt.show()
```

```python
#object oriented method
```

```python
fig = plt.figure()
axes = fig.add_axes([1,1,0.7,.7])
axes.plot(x,y)
plt.show()
```

```python
fig1 = plt.figure()
axis1 = fig1.add_axes([0.2,0.2,0.9,0.9])
axis2 = fig1.add_axes([0.3,0.6,0.4,0.4])
axis1.plot(x,y)
axis1.set_title('larger plot')

axis2.plot(y,x)
axis2.set_title('smaller plot')
plt.show()
```

```python
#subplot using oops
```

```python
figa,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout() #avoids overlapping
axes[0].set_title('first plot')
axes[0].plot(x,y)

axes[1].set_title('second plot')
axes[1].plot(y,x)
plt.show()
```

```python
#figure size
```

```python
figb,axes1 = plt.subplots(nrows=1,ncols=2,figsize=(8,5))
plt.tight_layout()
axes1[0].plot(x,y)
axes1[1].plot(x,x*2)
plt.show()
```

```python
figc = plt.figure()
ax = figc.add_axes([0,0,1,1])
ax.plot(x,y,label='xplot')
ax.plot(x,x*2,label='yplot')
ax.legend(loc=0)
plt.show()
```

```python
#plot appearence
```

```python
figd = plt.figure()
ax1 = figd.add_axes([0,0,1,1])
ax1.plot(x,y,linewidth=2,linestyle='--',alpha=0.8,marker='o',markerfacecolor='black',markersize=10)
ax1.grid(linestyle='--',color='green',alpha=0.5)
ax1.set_title('plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('y-axis')
plt.show()
```

```python
figd = plt.figure()
ax1 = figd.add_axes([0,0,1,1])
ax1.plot(x,y,linewidth=2,linestyle='--',marker='o')
ax1.set_xlim([0,1])
ax1.set_ylim([0,2])
plt.show()
```
