```python
from matplotlib import pyplot as plt
v = float(input("Enter the di/dt : "))
a = float(input("Enter the d^2i/dt^2 : "))
vin = float(input("Enter Initial voltage : "))
res = float(input("Enter the consolidated resistance : "))
time = int(input("Enter the time of observation : "))
vol = []
time = [x for x in range(time)]
for t in time :
  v += a*t
  vin += v*t*res
  vol.append(vin)
```

```python
plt.figure(figsize = (8,8))
plt.title("Transient Analysis")
#sns.scatter(time,vol)
plt.xlabel("Time(secs)")
plt.ylabel("Voltage(V)")
plt.plot(time,vol)
plt.show()
print(vol)
```
